# Installation of NEDAS and optional dependencies

More detailed guide can be found at [NEDAS documentation](https://nedas.readthedocs.io)

## Install via pip

NEDAS is available from the PyPI platform. To install the latested version:

```python
%pip install NEDAS
```

## Optional dependencies

- Numba provides JIT compilation of python function to speed it up during runtime

```python
%pip install numba
```

- pyFFTW provides an alternative implementation of FFT

```python
%pip install pyFFTW
```
